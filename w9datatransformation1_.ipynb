```python
# Load the 'w9.xlsx' Excel file into a pandas DataFrame and display its head
import pandas as pd
income=pd.read_excel('incomedataset.xlsx')
display(income.head())
```

```python
# Perform one-hot encoding on the 'income' DataFrame and display its head
e_income=pd.get_dummies(income,dtype=int)
display(e_income.head())
```

```python
#deserializing
#Create 5 bins for the 'age' column and display them
import numpy as np
bins =np.linspace(max(e_income["age"]),min(e_income["age"]),5)#no of groupings
bins
```

```python
#Assign each 'age' value to a bin and create a new 'b_age' column
#binning or discretization. It categorizes continuous numerical data
e_income["b_age"]=np.digitize(e_income["age"],bins=bins)
#display(e_income)
print(e_income)
```

```python
e_income=e_income.drop(columns="age")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Drop columns that are mostly NaN (the 'Unnamed' columns)
X = e_income.drop(columns=["income_ <=50K", "income_ >50K"])
y = e_income["income_ <=50K"]

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

#train the data
liReg=LogisticRegression().fit(X_train,y_train)#specifies which algorithm to use and trains
print("Training accuracy" ,liReg.score(X_train,y_train) )
print("test accuracy" ,liReg.score(X_test,y_test) )
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = e_income.drop(columns=["income_ <=50K", "income_ >50K"])
y = e_income["income_ <=50K"]
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)


# Initialize and train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

# Print the training and test accuracy for the Decision Tree model
print("Decision Tree Training accuracy:", dt_classifier.score(X_train, y_train))
print("Decision Tree Test accuracy:", dt_classifier.score(X_test, y_test))
```
