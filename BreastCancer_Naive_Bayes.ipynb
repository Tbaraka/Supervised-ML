```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
import numpy as np

# using real datasets
from sklearn.datasets import load_breast_cancer
breastcancer = load_breast_cancer()

# Splitting dataset
X_train, X_test, y_train, y_test = train_test_split(breastcancer.data, breastcancer.target, random_state=66)

# create/train model â€” FIXED HERE
NB = GaussianNB()  # Create instance
NB.fit(X_train, y_train)  # Then fit

print("Accuracy:", NB.score(X_test, y_test))
print("Predictions:", NB.predict(X_test))
```
