```python
#bias is due to use introducing a number it is not calculated like in y intercept
#there are more functions to use over twenty
#Activation Funcion
#epocs
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
X_train,Xtest,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)
mlp=MLPClassifier(random_state=42,hidden_layer_sizes=[100,50],activation="tanh",solver ="lbfgs",max_iter=10000,alpha =0.01)
mlp.fit(X_train,y_train)
print("Training score",mlp.score(X_train,y_train))
print("Testing Score ",mlp.score(Xtest,y_test))
```
