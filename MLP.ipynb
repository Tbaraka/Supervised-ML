```python
#bias is due to use introducing a number it is not calculated like in y intercept
#there are more functions to use over twenty
#Activation Funcion
#epocs
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris =load_iris()

# Initialize StandardScaler for feature scaling
sc=StandardScaler()

# Split the dataset into training and testing sets
X_train,Xtest,y_train,y_test=train_test_split(iris.data,iris.target,random_state=0)

# Scale the training and testing features
X_train=sc.fit_transform(X_train)
Xtest=sc.transform(Xtest)

# Initialize and train a Multi-layer Perceptron Classifier
# hidden_layer_sizes defines the architecture of the hidden layers
# max_iter sets the maximum number of iterations for the solver
# alpha is the L2 regularization parameter
mlp=MLPClassifier(random_state=66,hidden_layer_sizes=[5,10],max_iter=100000,alpha =0.01).fit(X_train,y_train)

# Print the training and testing scores (accuracy)
print("Training score",mlp.score(X_train,y_train))
print("Testing Score ",mlp.score(Xtest,y_test))
```
