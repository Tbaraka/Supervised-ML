```python
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn .model_selection import train_test_split
cancer=load_breast_cancer()
x_train,x_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)
model=SVC(kernel="rbf",C=1,gamma=0.03).fit(x_train,y_train)
print("Training score",model.score(x_train,y_train))
print("Test score",model.score(x_test,y_test))
```

```python
#scaling
from sklearn .preprocessing import MinMaxScaler
from sklearn.datasets import load_breast_cancer
from sklearn.svm import SVC
from sklearn .model_selection import train_test_split
from sklearn .preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cancer=load_breast_cancer()
x_train,x_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)
#we only scale the feature not the label
xtrain_scale=scaler.fit_transform(x_train)
#.fit is for training the model so we use transfrom
xtest_scale=scaler.transform(x_test)
model=SVC(kernel="rbf",C=1000,gamma=0.001).fit(xtrain_scale,y_train)
print("Training score",model.score(xtrain_scale,y_train))
print("Test score",model.score(xtest_scale,y_test))
```
