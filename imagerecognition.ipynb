```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler


digits=load_digits()

# Initialize StandardScaler for feature scaling
sc=StandardScaler()

# Split the dataset into training and testing sets
X_train,Xtest,y_train,y_test=train_test_split(digits.data,digits.target,random_state=66)

# Scale the training and testing features
X_train=sc.fit_transform(X_train)
Xtest=sc.transform(Xtest)

# Initialize and train a Multi-layer Perceptron Classifier
# hidden_layer_sizes defines the architecture of the hidden layers: a single hidden layer with 5 neurons.
# max_iter sets the maximum number of iterations for the solver to 100,000.
# alpha is the L2 regularization parameter, set to 0.01, to prevent overfitting.
# solver="sgd" specifies the Stochastic Gradient Descent optimizer.
# verbose=10 enables progress messages during training (printed every 10 iterations).
# early_stopping=True enables early stopping to terminate training if the validation score does not improve.
# n_iter_no_change=10 sets the number of iterations with no improvement after which training will be stopped.
# random_state=66 ensures reproducibility of the results.
mlp=MLPClassifier(random_state=0,max_iter=100000,alpha =0.01,solver="sgd",verbose=10,early_stopping=True,n_iter_no_change=10,hidden_layer_sizes=5).fit(X_train,y_train)

# Print the training and testing scores (accuracy)
print("Training score",mlp.score(X_train,y_train))
print("Testing Score ",mlp.score(Xtest,y_test))
```
