```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import brier_score_loss

# Load the Iris dataset
cancer =load_breast_cancer()


# Initialize StandardScaler for feature scaling
sc=StandardScaler()

# Split the dataset into training and testing sets
X_train,Xtest,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=0)

# Scale the training and testing features
X_train=sc.fit_transform(X_train)
Xtest=sc.transform(Xtest)
mlp=MLPClassifier(random_state=42,hidden_layer_sizes=[25],solver="sgd",verbose=10,early_stopping=True,n_iter_no_change=10,
                  max_iter=100000,alpha =0.01).fit(X_train,y_train)
p=mlp.predict_proba(Xtest)[:,0] # Corrected this line
b=brier_score_loss(y_test,p)


# Print the training and testing scores (accuracy)
print("Training score",mlp.score(X_train,y_train))
print("Testing Score ",mlp.score(Xtest,y_test))
#uncertainity
print("Predicted probabilities",mlp.predict_proba(Xtest[:6]))
print("Brier Score",b)
```
