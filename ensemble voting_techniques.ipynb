```python
#majority voting
#Hard Voting
#Each model votes for a class
#Class with most votes wins
#Used in classification

#Soft Voting
#Average predicted probabilities
#Choose class with highest average probability
#Used in classification (and sometimes regression)
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
##joinng all the algorithms
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
knn=KNeighborsClassifier(n_neighbors=3).fit(xtrain,ytrain)
dt= DecisionTreeClassifier().fit(xtrain,ytrain)
lr=LogisticRegression(max_iter=5000).fit(xtrain,ytrain) # Increased max_iter further to address ConvergenceWarning
ensemble=VotingClassifier(estimators =[('lr',lr),('knn',knn),('dt',dt)],voting="hard").fit(xtrain,ytrain)
print("lr training ",lr.score(xtrain,ytrain))
print("lr training ",lr.score(xtest,ytest))
print("knn training ",knn.score(xtrain,ytrain))
print("knn training ",knn.score(xtest,ytest))
print("dt training ",dt.score(xtrain,ytrain))
print("dt training ",dt.score(xtest,ytest))
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble training ",ensemble.score(xtest,ytest))
```

```python
#stacking voting
#Combine multiple base models using another machine learning model (meta-model or final estimator) to make predictions.
#their outputs become inputs to a final model -metalearner

from sklearn.datasets import load_breast_cancer
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
##joinng all the algorithms
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
knn=KNeighborsClassifier(n_neighbors=3).fit(xtrain,ytrain)
dt= DecisionTreeClassifier().fit(xtrain,ytrain)
lr=LogisticRegression(max_iter=10000).fit(xtrain,ytrain)
ensemble=StackingClassifier(estimators =[('lr',lr),('knn',knn),('dt',dt)],final_estimator=LogisticRegression()).fit(xtrain,ytrain)
print("lr training ",lr.score(xtrain,ytrain))
print("lr training ",lr.score(xtest,ytest))
print("knn training ",knn.score(xtrain,ytrain))
print("knn training ",knn.score(xtest,ytest))
print("dt training ",dt.score(xtrain,ytrain))
print("dt training ",dt.score(xtest,ytest))
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble training ",ensemble.score(xtest,ytest))
```

```python
#bagging-Bootstrap Aggregating
#Train same algorithm on different random subsets (with replacement)
#Reduces variance
#Makes predictions more stable
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
##joinng all the algorithms
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
dt= DecisionTreeClassifier().fit(xtrain,ytrain)
ensemble=BaggingClassifier(estimator =dt,n_estimators=5).fit(xtrain,ytrain)#This means 10 different Decision Tree models are trained on 10 bootstrapped samples of the data.
print("dt training ",dt.score(xtrain,ytrain))
print("dt test ",dt.score(xtest,ytest))
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble test ",ensemble.score(xtest,ytest))
```

```python
#bagging-randomforest
from sklearn.datasets import load_breast_cancer
##joinng all the algorithms
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
ensemble=RandomForestClassifier(n_estimators=100,max_features=10,max_depth=4).fit(xtrain,ytrain)#pruning
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble training ",ensemble.score(xtest,ytest))
```

```python
#boosting voting
#Models are built sequentially
#Each new model corrects previous mistakes
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
#joinng all the algorithms
from sklearn.ensemble import AdaBoostClassifier#adaptive boosting
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
dt= DecisionTreeClassifier().fit(xtrain,ytrain)
ensemble=AdaBoostClassifier(estimator =dt,n_estimators=10).fit(xtrain,ytrain)
print("dt training ",dt.score(xtrain,ytrain))
print("dt test ",dt.score(xtest,ytest))
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble test ",ensemble.score(xtest,ytest))
```

```python
#gradientboosting
from sklearn.datasets import load_breast_cancer
#joinng all the algorithms
from sklearn.ensemble import  GradientBoostingClassifier
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
dt= DecisionTreeClassifier().fit(xtrain,ytrain)
ensemble=GradientBoostingClassifier().fit(xtrain,ytrain)
print("dt training ",dt.score(xtrain,ytrain))
print("dt test ",dt.score(xtest,ytest))
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble test ",ensemble.score(xtest,ytest))
```

```python
#gradientboosting
from sklearn.datasets import load_breast_cancer
#joinng all the algorithms
from sklearn.ensemble import  GradientBoostingClassifier
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
xtrain,xtest,ytrain,ytest=train_test_split(cancer.data,cancer.target,random_state=66)
dt= DecisionTreeClassifier().fit(xtrain,ytrain)
ensemble=GradientBoostingClassifier(random_state=0,max_depth=1,learning_rate=0).fit(xtrain,ytrain)
print("dt training ",dt.score(xtrain,ytrain))
print("dt test ",dt.score(xtest,ytest))
print("ensemble training ",ensemble.score(xtrain,ytrain))
print("ensemble test ",ensemble.score(xtest,ytest))
```
