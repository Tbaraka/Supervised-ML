```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier

# Load data
wine = load_wine()
X_train, X_test, y_train, y_test = train_test_split(
    wine.data, wine.target, random_state=0
)

# Scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Base models
svm = SVC(kernel='rbf', C=10, gamma=0.1, probability=True)
knn = KNeighborsClassifier(n_neighbors=5)

# Explicitly fit base models for individual scoring
svm.fit(X_train_scaled, y_train)
knn.fit(X_train_scaled, y_train)

# Meta-learner
meta_model = LogisticRegression()

# Stacking model
stack_model = StackingClassifier(
    estimators=[('svm', svm), ('knn', knn)],
    final_estimator=meta_model
)

# Train stacking model
stack_model.fit(X_train_scaled, y_train)

#
print("Training score:", svm.score(X_train_scaled, y_train))
print("Testing score: ", svm.score(X_test_scaled, y_test))
print("Training score:", knn.score(X_train_scaled, y_train))
print("Testing score: ", knn.score(X_test_scaled, y_test))
print("Stacking Training Score:", stack_model.score(X_train_scaled, y_train))
print("Stacking Testing Score: ", stack_model.score(X_test_scaled, y_test))
```
