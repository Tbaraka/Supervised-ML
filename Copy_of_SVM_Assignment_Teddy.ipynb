# Identifiers

These fields only describe test metadata and do not influence COVID diagnosis.

batch_date, test_name, swab_type

# Too Many Missing Values

These features contain insufficient data, making them unreliable for training an SVM model.

rapid_flu_results, rapid_strep_results, ctab, rhonchi, wheezes, days_since_symptom_onset, cough_severity, sob_severity

# Free Text

These fields are unstructured and cannot be directly processed by SVM without additional preprocessing.

cxr_findings, cxr_impression, cxr_label, cxr_link

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
# Load the dataset
df = pd.read_csv('coronavirusdataset.csv')
# Encode non-numeric columns so the model can understand them
le = LabelEncoder()
for col in df.select_dtypes(include=['object', 'bool']).columns:
    df[col] = le.fit_transform(df[col].astype(str))
# Fill any remaining missing values with the median of each column
df = df.fillna(df.median())

# X = everything except the result column
# y = the covid test result column (what we want to predict)
X = df.drop(columns=['covid19_test_results'])
y = df['covid19_test_results']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
# class_weight='balanced' handles the imbalance (very few positive cases in our data)
model = SVC(kernel='rbf', C=10000, gamma=0.001, class_weight='balanced').fit(X_train_scaled, y_train)
# Print the accuracy
print('Training accuracy:', model.score(X_train_scaled, y_train))
print('Testing accuracy: ', model.score(X_test_scaled,  y_test))
```
