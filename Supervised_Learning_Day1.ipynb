```python
from sklearn.model_selection import train_test_split#helps us select a model that splits our data into column .
from sklearn.neighbors import KNeighborsClassifier#ML algorithm (K-NN model )
import mglearn
X,y =mglearn.datasets.make_forge()
#spliting the datasets
X_train,X_test,y_train,y_test =train_test_split(X,y,random_state=0)#spliting the data in rows.random state is included for uniformity of the model
#training the model
knn =KNeighborsClassifier(n_neighbors=9).fit(X_train,y_train)#naming the model(knn)the model is the trainign model
print(f"{(knn.score(X_test,y_test)*100):.2f}% ")
```

```python
from sklearn.model_selection import train_test_split#helps us select a model that splits our data into column .
from sklearn.neighbors import KNeighborsClassifier#ML algorithm (K-NN model )
import mglearn
X,y =mglearn.datasets.make_forge()
#spliting the datasets
X_train,X_test,y_train,y_test =train_test_split(X,y,random_state=100)#spliting the data in rows.random state is included for uniformity of the model
#training the model
knn =KNeighborsClassifier(n_neighbors=9).fit(X_train,y_train)#naming the model(knn)the model is the trainign model
print(knn.score(X_test,y_test))
print(knn.predict(X_test))
```

```python
#using real datasets
from sklearn.datasets import load_iris
iris=load_iris()
#spliting the datasets
X_train,X_test,y_train,y_test =train_test_split(iris.data,iris.target,random_state=3)
#training the model
knn =KNeighborsClassifier(n_neighbors=3).fit(X_train,y_train)
print(knn.score(X_test,y_test))
#print(iris.data)#features
#print(iris.target)#classes
```
